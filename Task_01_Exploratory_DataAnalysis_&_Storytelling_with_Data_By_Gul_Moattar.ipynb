### **1: DATASET**

Libraries used

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**IMPORTRING THE DATASET IN DATAFRAME.**

```python
df = pd.read_csv('/content/rest_delhi1.csv')
df
```

### **2: Clean and prepare the data**

```python
# To show the columns
df.columns
```

```python
# To show the number of rows in a dataset
df.shape
```

```python
# To see the info of dataset
df.info()
```

## Handle missing and / null values.

```python
# To remove the null values
# There is no null values in dataset.
df.isnull().sum()
df
```

## Remove duplicates

```python
#out of 912 rows here we got 764 rows totally 148 duplicates are removed
#to make this permanent change use inplace = True
df.drop_duplicates(inplace = True)
df
```

## Rename columns for clarity


```python
# To make it permanent change use inplace = True and rename the column names
df.rename(columns ={'num_of_rating':'ranking','price_for_two':'cost', 'name' : 'Names'},inplace = True)
df
```

## Correct datatypes

```python
df.dtypes
```

```python
df['Names'].unique()
```

```python
# Confirm 'rating' column is numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the average rating
location_rating = df.groupby('location')['rating'].mean().sort_values(ascending=False)

# Clean 'rating' column
df['rating'] = df['rating'].replace('--', np.nan)
df['rating'] = pd.to_numeric(df['rating'])

# Clean 'cost' column
df['cost'] = df['cost'].astype(str).str.replace('₹', '').str.replace(' ', '')
df['cost'] = pd.to_numeric(df['cost'])

# Remove rows with NaN values in 'rating' or 'cost' after cleaning
df.dropna(subset=['rating', 'cost'], inplace=True)

df.sort_values(by = 'rating',ascending = False)  #here data is sorted according to the rating of the cuisine
df.head(10)
```

### **4: Create visualizations**

# **1: Bar Graph : cuisine vs num of restaurants**

```python
df['cuisine'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Cuisines in Delhi Restaurants")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **2: line chart : location vs rating**

```python
# Select top N locations for better readability
top_n = 10
top_locations = location_rating.head(top_n)

plt.figure(figsize=(12, 6))
plt.plot(top_locations.index, top_locations.values, marker='o', linestyle='-')
plt.title(f'Average Rating by Location (Top {top_n})')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## **3: Pie chart : cuisines vs num of restaraunts**

```python
cuisine =df['cuisine'].value_counts().head(10)
plt.figure(figsize=(10, 8))  # Optional: Adjust figure size for better readability
plt.pie(cuisine,labels = cuisine.index,autopct = '%1.1f%%')
plt.title("cuisine vs num of restrauants")
plt.show()
```

## **4: Scatter chart: Names vs rating**

```python
plt.scatter(df['cost'], df['rating'], color='green', alpha=0.5)
plt.title("cost vs rating")
plt.xlabel("Names")
plt.ylabel("rating")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### **5: Histogram Graph**

```python
 # histogram
plt.hist2d(df['cost'], df['rating'], bins=(50, 50), cmap=plt.cm.plasma)
plt.colorbar()
plt.title("cost vs rating")
plt.xlabel("cost")
plt.ylabel("rating")
plt.show()
```

### **6: Heatmap chart**

```python
heatmap_data = df[['rating', 'cost']]

# Calculate the correlation matrix
correlation_matrix = heatmap_data.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Rating and Cost Correlation')
plt.show()
```
